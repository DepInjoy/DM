## 随机漫步

```python
from random import choice
import matplotlib.pyplot as plt

%matplotlib inline

class RandomWalk():
    def __init__(self, num_point = 5000):
        # 初始化点数
        self.num_point = num_point
        # 初始化开始的位置
        self.x = [0]
        self.y = [0]
        
    def fill_walk(self):
        while len(self.x) < self.num_point:
            # 计算X方向的步进，其中步进由方向和步长决定
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance
            
            # 同样地，计算y方向的步进
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance
            
            # 拒绝原地踏步
            if x_step == 0 or y_step == 0:
                continue
            
            # 计算下一个点的x和y的位置
            next_x = self.x[-1] + x_step
            next_y = self.y[-1] + y_step
            self.x.append(next_x)
            self.y.append(next_y)
            
```

```python
random_worker = RandomWalk()
random_worker.fill_walk()
plt.figure(figsize=(15, 5))
aes = plt.scatter(random_worker.x, random_worker.y, s = 15, 
                  c = list(range(random_worker.num_point)), cmap=plt.cm.Blues, 
                  edgecolors=None)
# 突出起点和终点的位置
plt.scatter(random_worker.x[0], random_worker.y[0], c = "yellow", s = 100)
plt.scatter(random_worker.x[-1], random_worker.y[-1], c = "red", s = 100)

# 隐藏坐标轴
aes.axes.get_xaxis().set_visible(False)
aes.axes.get_yaxis().set_visible(False)
plt.show()
```

## 模拟筛子

```python
from random import randint
import seaborn as sns

class Die():
    def __init__(self, num_sides = 6):
        '''
        初始化筛子的面数，默认为6
        '''
        self.num_sides = num_sides
    
    def roll(self):
        # 返回1到筛子面数之间的一个随机值
        return randint(1, self.num_sides)
```

```python
die = Die()
result = []

# 产生500次掷骰子的结果
for i in range(100):
    result.append(die.roll())
```

```python
# 可视化分析
sns.countplot(result, color=sns.palettes.color_palette()[0])
plt.xlabel("Die's Result")
plt.ylabel("Count")
plt.title("The frequence of die's result");
```
